## **Общая информация**

**NumPy** (Numerical Python Extensions - «Числовые расширения Python») – библиотека языка Python, добавляющая поддержку больших многомерных массивов и матриц и очень быстрые математические функции для операций с этими массивами. 

Ссылка на документацию библиотеки: https://numpy.org/doc/1.21/

*   Библиотека **Numpy** представляет собой хорошо-оптимизированный код на С и Fortran для работы с многомерными массивами.
*   Код библиотеки есть в свободном доступе на GitHub https://github.com/numpy/numpy.

**NumPy** включает в себя набор пакетов для решения специализированных задач, например:

*   `numpy.linalg` – реализует операции линейной алгебры;
*   `numpy.random` – реализует функции для работы со случайными величинами;
*   `numpy.fft` – реализует прямое и обратное преобразование Фурье.

### Представление массива (поверхностная копия, view). метод **`view()`**

Бывают случаи, когда копия массива не нужна, а нужен тот же самый массив, но с другими размерами – другое представление исходного массива.

Для таких нужд **NumPy** предоставляет метод **ndarray.view()**. Этот метод создает новый объект массива, который просматривает данные исходного массива, но изменение размеров одного массива не влечёт изменение размеров другого.

Та же область памяти, но данные представляются в другой форме.

```python
a = np.array([[1, 1, 3.1],
              [4, 9, 1.1]])

print(f'Массив а:\n {a}')
print(f'Массив а:\n {a.strides}') 
```

```python
b = a.view() # создаётся новый объект – массив, являющийся представлением тех же данных

print(f'b и a - это один объект: {b is a}')
print(f'Форма массива а: {a.shape}')
print(f'Форма массива b: {b.shape}')

# изменим форму массива b
b = b.reshape(1, 6) # изменение формы массива b, a.shape не изменится

print(f'\nФорма массива а после изменения формы массива b: {a.shape}')
print(f'Массив а:\n {a}')
print(f'Форма массива b после изменения формы массива b: {b.shape}')
print(f'Массив b:\n {b}')

# изменим какое-нибудь значение в массиве b
print(f'\nЗначение a[0, 0]: {a[0, 0]}')
b[0, 0] = -1 # это значение изменится и в массиве a
print('Значение b[0, 0] заменили на -1')
print(f'Значение a[0, 0]: {a[0, 0]}')
```

### Изменение формы массива. методы **`reshape()`**, **`ravel()`**, **`flatten()`**



```python
a = np.array([[1, 1, 3.1], [4, 9, 1.1]])

print(f'strides массива a: {a.strides}') # кортеж байтов для шага в каждом измерении при обходе массива
print(f'тип данных массива a: {a.dtype}')
```

**`ndarray.reshape()`** - меняет форму массива на заданную (создаётся представление)

```python
# изменим форму массива
a = np.zeros((10, 3))
print(f'массив а:\n {a}')
print(f'Форма массива a: {a.shape}')

b = a.reshape(2, 15) # указывает на те же данные, что и массив a, но создаётся новый объект - отличается атрибутами shape и strides
# b = a.reshape(2, -1) # альтернативная запись строчки выше
print(f'\nb и a - это один объект: {a is b}') # разные объекты
print(f'Форма массива a: {a.shape}')
print(f'Форма массива b: {b.shape}')

# изменим какое-нибудь значение в массиве b
print(f'\nЗначение a[0, 0]: {a[0, 0]}')
b[0, 0] = -1 # это значение изменится и в массиве a
print('Значение b[0, 0] заменили на -1')
print(f'Значение a[0, 0]: {a[0, 0]}')
```

**`ndarray.flatten()`** – разворачивает многомерный массив в строку (создаётся глубокая копия)

```python
b = a.flatten()  # создана глубокая копия
b[0] = -10 # данные a не изменятся

print(f'массив a:\n {a}')
print(f'Форма массива b: {b.shape}')
```

**`ndarray.ravel()`** – разворачивает многомерный массив в строку (создаётся представление)

Это аналог `ndarray.reshape(-1)`

```python
b
```

```python
b = a.ravel()  # создано представление (другой способ выполнить эту же операцию a.reshape(-1))
b[0] = -5 # данные a изменятся

print(f'массив a:\n {a}')
```

## **Индексация, срезы**

*   Операции индексирования, срезов и итераций выполняются очень схожим образом с обычными списками и другими последовательностями Python.

*   Доступ к элементам массива осуществляется по целочисленным индексами, начинается отсчёт с 0.

*   При использовании неполного набора индексов, недостающие индексы неявно заменяются списком всех возможных индексов вдоль соответствующей оси. Сделать это явным образом можно, поставив ":".

*   Индексы могут принимать отрицательные целые значения. В этом случае отсчет ведется от конца массива.

*   Индексы указываются в квадратных скобках.

    Формат: `[первый индекс, последний индекс, шаг]`

    Значения по умолчанию: `[0, len(массив), 1]`

    Элемент с последним индексом не включается.

*   Срез – это представление

*   Получать значение в многомерном массиве по индексу так a[0][0] – плохо, a[0, 0] – хорошо

```python
a = np.arange(15)

print(f'массив a:\n {a}')
```

```python
# пример обращения по индексу и взятия среза
b = a[0:2] # это представление
b[0] = 100 # значение в a изменится

print(f'массив a:\n {a}')
```

### Булевая индексация

```python
a = np.array([[5, -1, 2, 3, 1],
             [0, -1, -6, 5, 5]])

print(f'массив a:\n {a}')
```

```python
# зададим маску
mask = a < 0

print(mask)
```

```python
# фильтрация
a[mask] # будет возвращён массив тех элементов a, для которых значение маски True
```

```python
# найдём элементы массива, кратные или 3 или 5
mask3 = a % 3 == 0
mask5 = a % 5 == 0

a[np.logical_or(mask3, mask5)] # или a[mask3 | mask5]
```

```python
# получим строки, в которых есть хотя бы один 0
mask = (a == 0).any(axis=1)
a[mask] 
```

```python
# получим столбцы, в которых число положительных элементов больше числа отрицательных
mask = (a > 0).sum(axis=0) > (a < 0).sum(axis=0) # метод ndarray.sum() будет рассмотрен далее
a[:, mask] 
```

**`numpy.where()`** – получение индексов элементов массива, которые удовлетворяют маске

## **Фиктивные оси**

Фиктивная ось – ось, размер которой равен 1.

### Агрегирующие операции

Агрегирующие операции используются для получения обобщающих значений по данным (получают по ним единственное значение). К  ним относят различные операции над выборкой, например, получение числа элементов, получение минимального, максимального и среднего значения в выборке, а также суммирование значений

*   В библиотеке для многих таких операций есть специальные методы:
`a.min()`,` a.max()`, `a.argmax()`, `a.sum()`,` a.prod()`, `a.mean()` и др.
и функции: `np.min(a)`, `np.max(a)`, `np.argmax(a)`, `np.sum(a)`, `np.prod(a)`, `np.mean(a)` и др.

*   Циклы использовать не нужно!!!

*   Не рекомендуется также использовать функции Python: `min(a)`, `max(a)`, `sum(a)` и др. (и то это сработает только для одномерных массивов) - произойдет приведение массива к list, далее выполнится проход в цикле по list;
Если попытаться применить эти функции к многомерному массиву, то будет ошибка, т.к. преобразование к list не сможет выполниться

*   По умолчанию, агрегирующие операции применяются к массиву, как если бы он был списком чисел, независимо от его формы.

```python
a = np.array([[1, 2, 3], [-1, 44, 5]])
a.min() # меньшее число в массиве
```

*   Для многомерных массивов каждая из функций может принять дополнительный аргумент `axis` и в зависимости от его значения выполнять функции по определенной оси, помещая результаты исполнения в массив.

```python
a.min(axis=0) # меньшее число в каждом столбце
```

```python
a.min(axis=1) # меньшее число в каждой строке
```

## **Сравнение массивов `numpy.ndarray` со списками `list` по эффективности**

```python
import time
```

```python
n = 10000000
```

```python
# найдём сумму элементов списка Python (если элементов много, то считаться будет долго)
def sum_lst(lst):
  sum = 0
  for i in lst:
    sum += i
  return sum

lst = [i for i in range(n)]
t0 = time.time()
s = sum_lst(lst)
t1 = time.time() - t0

print(f'Сумма из {n} элементов равна: {s}')
print(f'Время выполнения суммирования: {t1} сек')
```

```python
# найдём сумму элементов массива numpy.ndarray (большое число элементов, рассчитывается быстрее, чем в примере выше)
a = np.arange(n)
t0 = time.time()
s = a.sum()
t1 = time.time() - t0

print(f'Сумма из {n} элементов равна: {s}')
print(f'Время выполнения суммирования: {t1} сек')
```

### Операции над несколькими массивами. Транслирование массивов (Broadcasting).

*   Массивы должны быть одинаковых размеров, либо одна из размерностей должна быть == 1 (Если число элементов в размерности не совпадают и одна из них == 1, то та размерность, что равна 1, будет повторяться, пока размерности не совпадут).
* При работе с массивами разных типов, тип результирующего массива будет соответствовать более общему или более точному типу.

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])
```

```python
b = np.array([[1, -1, 1]], dtype=float) # [[1, -1, 1], [1, -1, 1]]
```

```python
print(f"Массив а: {a}, тип: {a.dtype}")
print(f"Массив b: {b}, тип: {b.dtype}")
```

```python
a + b # np.add(a, b), поэлементное сложение, результат типа более общего типа из двух - float
```

```python
a * b # np.multiply(a, b), поэлементное умножение
```

Выше при арифметических операциях с массивами, например, при сложении и умножении, перемножались массивы одинаковой формы.


Broadcasting

```python
# самый простой случай
a = np.array([1, 2, 3])
b = np.array([2, 2, 2])

print(a * b)
```

Произошло поэлементное умножение, все элементы массива  a  умножились на  2 . Но можно сделать проще, просто умножив массив на 2 .

```python
# умножение массива на число
print(a * 2)
```

На самом деле поведение будет аналогичным, если умножить одномерный массив на массив длины  1.

```python
# умножение массивов разных длин
b = np.array([2])
print(a * b)
```

В этом случае работает транслирование массивов (broadcasting). Один массив "растягивается" так, чтобы повторить форму другого.
image.png

Такой же эффект работает и для многомерных массивов. Если по какому-то измерению размер у одного массива равен 1, а у другого – произвольный, то по этому измерению может произойти "растяжение". Таким образом, массивы можно умножать друг на друга, если в измерениях, где они по размеру не совпадают, хотя бы у одного размер 1. Для других поэлементных операций правило аналогично.

Размерности сопоставляются справа налево. Если их количество не совпадает, то массивы меньшей размерности сначала дополняются слева размерностями 1. Например, при сложении массива размера  4×3  с массивом размера  3  последний сначала преобразуется в массив размера  1×3 .

```python
a = np.array([[ 0,  0,  0],
              [10, 10, 10],
              [20, 20, 20],
              [30, 30, 30]])

b = np.array([0, 1, 2])

print(a + b)
```

Пользоваться транслированием массивов нужно с осторожностью. Многократное копирование массива при растяжении может привести к неэффективной работе с памятью. Особенно за этим приходится следить при работе с GPU.

## Векторная и матричная математика

**NumPy** обеспечивает много функций для работы с векторами и матрицами

```python
# скалярное произведение векторов
a = np.array([1, 2, 3], float)
b = np.array([0, 1, 1], float)

np.dot(a, b)
```

```python
# нахождение определителя матрицы
a = np.array([[4, 2, 0], [9, 3, 7], [1, 2, 1]], float)

np.linalg.det(a)
```

```python
# нахождение обратной матрицы
a = np.array([[0, 1], [2, 3]])
a1 = np.linalg.inv(a)

print(a1)
```

```python
# нахождение собственного вектора и собственных значение матрицы
vals, vecs = np.linalg.eig(a)

print(f'Собственный вектор:\n {vecs}')
print(f'\nСобственные значения:\n {vals}')
```
